```python
#Visualise the missingness for the data
# Calculate the % of missing values for each column

missing_percentage = (pop.isnull().sum()/len(pop)) * 100

# Create a bar plot of the missing value
# sns is calling the programme
plt.figure(figsize = (12,6))
sns.barplot(x = missing_percentage.index, y = missing_percentage.values)
plt.xticks(rotation = 90)
plt.xlabel('Columns')
plt.ylabel('Percentage of MIssing Values')
plt.title('Missing Values in Populatuon Data')
plt.show()
```
