Ingesting image data & basic prop of the image


```python
from PIL import Image, ExifTags
import matplotlib.pyplot as plt

chaplin = Image.open('/content/Chaplin.png')

chaplin #prints the image
```

```python
# To print image format, size & mode
# image.size returns dimensions (width,height)
# image.mode returns colour e.g 'RGB','L' for greyscale, 'CMYK' for pre-press images
def basic_info(image):
    y = print(f"The picture format is {image.format}, the size of the pictutre is {image.size}, and the mode is {image.mode}")

    return y
```

```python
# if image was not read from the file, its set to None
basic_info(chaplin)
```

```python
snowboarder = Image.open('/content/snowboarder.jpg')
basic_info(snowboarder)
```
